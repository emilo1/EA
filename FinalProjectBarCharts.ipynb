```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
csv_path = "ESportsGames.csv"

games_df = pd.read_csv(csv_path, encoding = "utf-8")
games_df.head()
```

```python
grouped_by_genre_df = games_df.groupby("genre")
grouped_by_genre_df["genre"].count()
```

```python
game_groups = grouped_by_genre_df["genre"].count()

genre_chart = plt.bar(list(game_groups.index), list(game_groups.values))

plt.xlabel("Genre")
plt.ylabel("Games")
plt.xticks(rotation = "vertical")


plt.show()
# plt.tight_layout()
# plt.savefig("FinalProjectBarGraph1")
```

```python



player_groups = grouped_by_genre_df["totalplayers"].sum()

player_chart = plt.bar(list(player_groups.index), list(player_groups.values))

plt.xlabel("Genre")
plt.ylabel("Number of Players")
plt.xticks(rotation = "vertical")


plt.show()
# plt.tight_layout()
# plt.savefig("FinalProjectBarGraph2")
```

```python

fig, axes = plt.subplots(1, 2, figsize = (15, 5))






axes[0].bar(list(game_groups.index), list(game_groups.values))
axes[0].set_xlabel("Genre")
axes[0].set_ylabel("Games")
plt.setp(axes[0].get_xticklabels(), Rotation = 90)


axes[1].bar(list(game_groups.index), list(game_groups.values))
axes[1].set_xlabel("Genre")
axes[1].set_ylabel("Players")
plt.setp(axes[1].get_xticklabels(), Rotation = 90)



matplotlibsubplots = plt.show()

matplotlibsubplots





```
