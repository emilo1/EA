```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_path = "ESportsGames.csv"

games_df = pd.read_csv(csv_path, encoding = "utf-8")
games_df.head()
```

```python
grouped_by_genre_df = games_df.groupby("genre")
grouped_by_genre_df["genre"].count()
```

```python
game_groups = grouped_by_genre_df["genre"].count()

genre_chart = game_groups.plot(kind = "bar", title = "ESports Games by Genre")
genre_chart.set_xlabel("Genre")
genre_chart.set_ylabel("Number of Games")

plt.show()
plt.tight_layout()
plt.savefig("FinalProjectBarGraph1")
```
