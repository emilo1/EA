```python
%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
```

```python
twitch = "Twitch_user_data.csv"
twitch_df = pd.read_csv(twitch)
twitch_df
```

```python
tw_renamed_df = twitch_df.rename(columns={"Watch time(Minutes)":"Watch", "Stream time(minutes)": "Stream"})
tw_renamed_df
```

```python
tw_renamed_df.plot(kind="scatter", x="Followers", y="Followers gained", figsize=(20,20),
              title="Followers Vs. Followers Gained")
plt.show()
```

```python
average_peak = tw_renamed_df['Peak viewers'].mean
view_mean = tw_renamed_df['Views gained'].mean
viewer_mean = tw_renamed_df['Average viewers'].mean

print(average_peak)
print(view_mean)
print(viewer_mean)
```
